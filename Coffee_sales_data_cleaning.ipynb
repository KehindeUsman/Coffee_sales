```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv(r"C:\Users\Kehinde\Downloads\dirty_cafe_sales.csv")
data.head()
```

```python
data.info()
```

```python
#Checking for non-values
data.isna().sum()
```

```python
data['Transaction Date'].value_counts()
```

```python
data['Transaction Date'] = data['Transaction Date'].replace(['UNKNOWN', 'ERROR'], pd.NA)
```

```python
data['Transaction Date'].value_counts()
```

```python
data.isna().sum()
```

```python
data.dropna(subset=['Transaction Date'], inplace=True)
```

```python
#Converting date column to datetime data type
data['Transaction Date']= pd.to_datetime(data['Transaction Date'])
```

```python
data.info()
```

```python
data['Item'].value_counts()
```

```python
#Replace all ERROR in data['Item'] to UNKNOWN
data['Item'] = data['Item'].replace({'ERROR': 'UNKNOWN'})
```

```python
# Fill null values as UNKNOWN
data['Item'] = data['Item'].fillna('UNKNOWN')
```

```python
data.isna().sum()
```

```python
data['Location'].value_counts()
```

```python
#Replace all ERROR in data['Location'] to UNKNOWN
data['Location'] = data['Location'].replace({'ERROR': 'UNKNOWN'})
```

```python
data['Location'].value_counts()
```

```python
data.isna().sum()
```

```python
data['Location'] = data['Location'].fillna('UNKNOWN')
```

```python
data['Location'].value_counts()
```

```python
data['Payment Method'].value_counts()
```

```python
#Replace all ERROR in data['Payment Method'] to UNKNOWN
data['Payment Method'] = data['Payment Method'].replace({'ERROR': 'UNKNOWN'})
```

```python
data.isna().sum()
```

```python
data['Payment Method'] = data['Payment Method'].fillna('UNKNOWN')
```

```python
data.isna().sum()
```

```python
data['Quantity'].value_counts()
```

```python
data['Quantity'] = data['Quantity'].replace(['ERROR', 'UNKNOWN'], 1)
```

```python
data['Quantity'].value_counts()
```

```python
data['Price Per Unit'].value_counts()
```

```python
data['Price Per Unit'] = data['Price Per Unit'].replace(['ERROR', 'UNKNOWN'], 1.0)
```

```python
data['Quantity'] = pd.to_numeric(data['Quantity'], errors='coerce')
```

```python
data['Price Per Unit'] = pd.to_numeric(data['Price Per Unit'], errors='coerce')
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
#Fill with mean or median but since there isn't outliers we fill with mean 
data['Quantity'] = data['Quantity'].fillna(data['Quantity'].mean())
```

```python
#Fill with mean or median but since there isn't outliers we fill with mean 
data['Price Per Unit'] = data['Price Per Unit'].fillna(data['Price Per Unit'].mean())
```

```python
data.isna().sum()
```

```python
data['Total Spent'] = data['Quantity'] * data['Price Per Unit']
```

```python
data.isna().sum()
```

```python
data.duplicated().sum()
```

```python
#Saving DataFrame to CSV 
data.to_csv('cleaned_cafe_sales.csv', index=False)
```
